7\. Prime numbers sieve: compute the prime numbers in the 0-N (N=99 to start with) range with a sieve (mask).
  * Construct a shape (100,) boolean array, the mask
  * Identify the multiples of each number starting from 2 and set accordingly the corresponding mask element
  * Apply the mask to obtain an array of ordered prime numbers
  * Check the performances (timeit); how does it scale with N?
  * Implement the optimization suggested in the [sieve of Eratosthenes](https://en.wikipedia.org/wiki/Sieve_of_Eratosthenes)

````python
N=100
mask=np.ones((N,),bool)
p=2
mask[0]=mask[1]=False #escludo 0 e 1. Si potrebbe direttamente partire da 2, ma il testo del problema dice di usare un array di N elementi

def find_first_true_after(x,p) : #metodi come where e nonzero fanno una scansione di tutto l'array, ma a me basta solo il primo True dopo p
    for i,elem in enumerate(x) :#cfr https://stackoverflow.com/questions/432112/is-there-a-numpy-function-to-return-the-first-index-of-something-in-an-array quarta risposta
        if elem == True and i>p : return i #potrei usare np.ndenumerate(x) ma così i sarebbe un array di tipo (1,) mentre a me serve un intero (si potrebbe usare int(), chiaramente)
    return None #non so cosa convenga

while(p is not None) :
    #print(p)
    mask[2*p::p]=False #non voglio includere p stesso, quindi parto dal multiplo successivo (=2p) 
    p=find_first_true_after(mask,p) 

n=np.arange(0,N)
print(n[mask])
````

````python
#reimplemento il blocco di codice precedente come funzione di N, con e senza ottimizzazione
def find_first_true_after(x,p) : #copio qua questa funzione così questo blocco può essere eseguito autonomamente
    for i,elem in enumerate(x) :
        if elem == True and i>p : return i 
    return None 

def first_N_primes(N) :
    mask=np.ones((N,),bool)
    p=2
    mask[0]=mask[1]=False
    while(p is not None) :
        mask[2*p::p]=False
        p=find_first_true_after(mask,p)

    n=np.arange(0,N)
    return n[mask]

#l'ottimizzazione suggerita su wikipedia consiste nel partire da p^2 anziché da p(*2) e nel lavorare solo con numeri dispari e incrementi di 2*p (quest'ultima modifica può essere "nascosta", vedi sotto)

def first_N_primes_opt(N) :
    r=N%2
    if r == 0 : mask=np.ones((int(N/2)-1,),bool) #se N è pari i numeri da 0 a N-1 sono metà pari e metà dispari, quindi i numeri dispari sono N/2 - da cui però escludo uno
    else : mask=np.ones((N//2-1,),bool)#se N è dispari ed escludo l'ultimo numero (cioè N-1) allora siamo nel caso precedente, quindi (N-1)/2 numeri sono dispari ed altrettanti pari;    il numero che ho escluso è l'ultimo, N-1, e per definizione è pari, quindi complessivamente ho (N-1)/2=N/2-1/2=N//2 numeri dispari ed (N-1)/2+1 pari. Nuovamente parto da 3, cioè     escludo 1
    #in alternativa: da 1 a 100 ci sono 50 numeri pari e 50 dispari, da 0 a 99 altrettanti perché non ho toccato i numeri dispari e ho eliminato un numero pari (100) aggiungendone       un altro (0)

    p=3 #parto da 3, includo 2 a mano alla fine
    #ATTENZIONE: avendo eliminato i numeri pari e 1 non c'è più corrispondenza fra il numero primo p considerato e l'indice che ne dà la posizione nell'array!
    #in particolare: se d è un numero dispari precedentemente di indice d il nuovo indice può essere calcolato considerando quanto segue. 
    #la lista [0,..,d] può essere ottenuta mediante range(d+1), quindi ragionando come sopra in [0,..,d] ci sono (d+1)/2 numeri pari e altrettanti dispari; per usare solo i numeri       dispari a partire da 3 devo quindi eliminare (d+1)/2 numeri pari più il numero 1, cioè il nuovo indice è quello vecchio meno (d+1)/2+1, cioè d'=d-(d+3)/2

    
    while(p is not None) :
        #print(p)
        q=int(p-(p+3)/2)#nuovo indice di p
        q2=int((p**2)-(((p**2)+3)/2)) #mi serve il nuovo indice di p^2 come punto di partenza per quanto segue; il passo, invece, non va modificato poiché le distanze fra numeri restano invariate
        mask[q2::p]=False #non metto un passo di 2*p perché quel 2 serve solo se i numeri pari non vengono eliminati dalla lista
        q=find_first_true_after(mask,q)
        try : 
            p=int(2*(q+(3/2))) #trovo il numero primo corrispondente al nuovo indice q; tale numero serve per determinare il passo della prossima iterazione 
        except :
            p=None #serve questo costrutto perché non posso fare operazioni con None!

    if r == 0 : n=np.arange(3,N,2)
    else : n=np.arange(3,N-1,2)
    return np.concatenate((np.array([2]),n[mask]))

#NOTA: è evidente che se la funzione di cui sopra da un lato permetta di usare meno memoria (scartiamo a monte i numeri pari) dall'altro richieda però molti più conti, quindi non è detto che sia la cosa più conveniente. Scrivo allora un'altra funzione in cui NON scarto i numeri pari e ad ogni iterazione parto da p^2 e vado a passi di 2*p; è possibile che per N "piccoli" usare un po' più di memoria sia più conveniente che fare più calcoli anche se semplici! Confrontando queste due con timeit potremo vedere se convenga risparmiare memoria facendo meno conti o il contrario


def first_N_primes_opt_simple(N) :
    mask=np.ones((N,),bool)
    #per lavorare solo con i numeri dispari metto un False in corrispondenza di ogni numero pari diverso da 2 e parto da p=3
    mask[4::2] = False #parto da 4 perché 2 è primo
    mask[0]=mask[1]=False #0 e 1 non sono primi
    p=3
    while(p is not None) :
        mask[p**2::2*p]=False
        p=find_first_true_after(mask,p)

    n=np.arange(0,N)
    return n[mask]


l=100
#a=first_N_primes(l)
#print(a)  
#b=first_N_primes_opt(l)
#print(b)
#print(a==b)
#c=first_N_primes_opt_simple(l)
#print(c)
#print(a==c)

def test_prime_numbers(num) :
    print("time needed to compute the first %d prime numbers"%num)
    print("non-optimized function:")
    %timeit first_N_primes(num)
    print("memory-light optimized function:")
    %timeit first_N_primes_opt(num)
    print("memory-heavy optimized function:")
    %timeit first_N_primes_opt_simple(num)

for l in [100,1000,10000] :
    test_prime_numbers(l)
    print("\n")


#sembra che risparmiare memoria anche a costo di fare più conti sia meglio di circa un fattore 2 se l=100, l=1000 o l=10000. Sorprendentemente l'ultima funzione dà guadagni poco significativi rispetto alla prima! Solo la seconda funzione dà un guadagno di circa il 50%
````

8\. Diffusion using random walk

Consider a simple random walk process: at each step in time, a walker jumps right or left (+1 or -1) with equal probability. The goal is to find the typical distance from the origin of a random walker after a given amount of time. 
To do that, let's simulate many walkers and create a 2D array with each walker as a row and the actual time evolution as columns

  * Take 1000 walkers and let them walk for 200 steps
  * Use randint to create a 2D array of size walkers x steps with values -1 or 1
  * Build the actual walking distances for each walker (i.e. another 2D array "summing on each row")
  * Take the square of that 2D array (elementwise)
  * Compute the mean of the squared distances at each step (i.e. the mean along the columns)
  * Plot the average distances (sqrt(distance\*\*2)) as a function of time (step)
  
Did you get what you expected?

````python
#it's probably best to first generate all random numbers, which must then represent the step taken by each walker at each instant; then performing the appropriate sum along some dimension we get the actual final position=net distance travelled by each walker
N=1000 #n. of walkers
t=200#n. of steps
W=npr.randint(0,2,size=(N,t))
W[W==0]=-1#se qua sopra metto -1,2 verranno generati come numeri casuali -1,0,1 anziché -1 e 1; così evito il problema. Forse si può direttamente estrarre o -1 o 1
#print(W)
for i in range(1,t) :#escludo la prima colonna, l'unica che va lasciata stare; alle altre devo sommare tutte le precedenti, quindi per induzione mi basta sommare alla col. i la i-1
    W[:,i]+=W[:,i-1]
#print(W)
v=np.mean(W**2,axis=0)
%matplotlib inline
T=np.linspace(0,t,t)
plt.plot(T,np.sqrt(v));
plt.plot(T,np.sqrt(T));#sigma^2 should be linear in time i.e. sigma=sqrt(v) should be proportional to sqrt(t)
````

````python
import matplotlib.pyplot as plt
M=np.loadtxt("populations.txt",skiprows=1)
#print(M)
year=M[:,0]
hare=M[:,1]
lynx=M[:,2]
carrots=M[:,3]

%matplotlib inline
plt.plot(year,hare,label="hare pop.(t)");
plt.plot(year,lynx,label="lynx pop.(t)");
plt.plot(year,carrots,label="carrots(t)");
plt.legend();

mean=np.zeros((3,))
std=np.zeros((3,))
for i,j in enumerate([carrots,hare,lynx]) :
    mean[i]=np.mean(j)
    std[i]=np.std(j)
for i,j in enumerate(["carrots","hares","lynxes"]) :
    print("Average number of %s ="%j,mean[i])
    print("with std =",std[i])

corr_carrots_hares=np.corrcoef(M[:,0:2],rowvar=False)
corr_hares_lynxes=np.corrcoef(M[:,1:3],rowvar=False)
corr_carrots_lynxes=np.corrcoef(M[:,(0,2)],rowvar=False)
#corr_carrots_lynxes=np.corrcoef(np.array((M[:,0],M[:,2])))#anche così va bene
print("carrots-hares correlation =",corr_carrots_hares[0,1])#ottengo [[corr(x,x) corr(x,y)],[corr(x,y),corr(x,x)]] come output di np.corrcoef(), quindi seleziono il secondo argomento
print("hares-lynxes correlation =",corr_hares_lynxes[0,1])
print("carrots-lynxes correlation =",corr_carrots_lynxes[0,1])
````
